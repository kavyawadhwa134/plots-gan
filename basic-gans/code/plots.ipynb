```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt

sb.set()

data = pd.read_csv('loss_logs.csv')

# Check the structure of the DataFrame
print("DataFrame shape:", data.shape)
print("DataFrame head:\n", data.head())

# Ensure that the necessary columns exist
required_columns = ['Iteration', 'Discriminator Loss', 'Generator Loss']
if not all(col in data.columns for col in required_columns):
    raise ValueError(f"The data must contain the following columns: {', '.join(required_columns)}")

# Plot the Discriminator and Generator Loss
plt.plot(data['Iteration'], data['Discriminator Loss'], label='Discriminator Loss')
plt.plot(data['Iteration'], data['Generator Loss'], label='Generator Loss')

plt.legend()
plt.title('Training Losses')
plt.tight_layout()
plt.savefig('../plots/training_loss.png')
plt.show()
```
