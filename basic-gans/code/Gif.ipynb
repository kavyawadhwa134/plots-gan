```python
import imageio.v2 as imageio

from PIL import Image
import imageio
import os

def create_gif(ipath, output_path, frames=11, step=1000, fps=2):
    images = []
    for i in range(frames):
        img_path = ipath % (i * step)
        if os.path.exists(img_path):
            images.append(imageio.imread(img_path))
        else:
            print(f"Warning: File not found: {img_path}")
    if images:
        imageio.mimsave(output_path, images, fps=fps)
    else:
        print(f"No images found to create {output_path}")

# Create GIFs
create_gif("../plots/features/feature_transform_%d.png", "../images/feature_transform.gif")
create_gif("../plots/features/feature_transform_centroid_%d.png", "../images/feature_transform_centroid.gif")
create_gif("../plots/iterations/iteration_%d.png", "../images/iterations.gif")
```

```python
import imageio.v2 as imageio
import os

def create_gif(ipath, output_path, frames=11, step=1000, fps=2):
    images = []
    for i in range(frames):
        img_path = ipath % (i * step)
        if os.path.exists(img_path):
            images.append(imageio.imread(img_path))
        else:
            print(f"Warning: File not found: {img_path}")
    if images:
        imageio.mimsave(output_path, images, fps=fps)
        print(f"GIF saved at: {output_path}")
    else:
        print(f"No images found to create {output_path}")

# Parameters for GIF creation
input_path = "../plots/iterations/iteration_%d.png"  # Path to the iteration images
output_path = "../images/iterations.gif"              # Path to save the GIF
frames = 13                                           # Total number of frames to include in the GIF
step = 1000                                          # Step size for image naming
fps = 2.5                                              # Frames per second for the GIF

# Create the GIF
create_gif(input_path, output_path, frames=frames, step=step, fps=fps)
```
