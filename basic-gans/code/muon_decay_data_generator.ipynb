```python
import numpy as np

# Parameters for muon decay
N0 = 1000  # Initial number of muons
decay_constant = 0.1  # Decay constant (you can adjust this value)
time_steps = np.arange(0, 20, 0.1)  # Time from 0 to 20 seconds in 0.1 second intervals

# Calculate the number of muons remaining at each time step
N_t = N0 * np.exp(-decay_constant * time_steps)

# Prepare data for saving
data = np.column_stack((time_steps, N_t))

# Save the data to a text file
output_file = 'muon_decay_data.txt'
np.savetxt(output_file, data, header='Time (s)  Number of Muons', fmt='%.2f %.2f')

print(f"Dataset generated and saved to {output_file}.")
```
